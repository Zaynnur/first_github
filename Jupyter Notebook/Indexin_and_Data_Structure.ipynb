## Dictionaries
- An unordered collection of elements
- Key and Value
- curly braccket { }
- mutateable/Change the value

```python
#Food and their prices
food1 = {"Samosa":110, "Pakora":100,"Raita":60,"Salaad":50,"Chicken Rolls":80}
food1
```

```python
type(food1)
```

```python
# Extract Data
keys1 = food1.keys()
keys1
```

```python
value1 = food1.values()
value1
```

```python
food1["Tikki"]=50
food1
```

```python
# Update the Values
food1["Tikki"]=60
food1
```

```python
food2 = {"Dates":150, "Cholates":200,"Swayyan":100}
food2
```

```python
# Concatinate
food1.update(food2)
food1
```
